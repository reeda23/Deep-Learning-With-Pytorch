# Activation function 
Activation functions apply a non-linear transformation and decide whether a neuron should be activated or not.

**Why we use activation functions?** <br>
So why we use activation functions and why only a linear transformation is not good enough? <br>

Typically we have a linear layer in our network that applies a linear transformation where it multiplies input with some weights and add some bias and delivers the output.

linear transformation.PNG <br>
Suppose we if don't have activation functions in between the we would have only linear transformations after each other so our whole network from inpu to output is essentially just a linear regression model and this linear model is not suited for complex task.

So the conclusion is that with nonlinear transformations in between our network can learn better and perform more complex tasks. <br>

# Leaky ReLU Function


For negative numbers it will multiply input with the samll value e.g 0.001. For positive it outputs, the input as it is. This is improved version of ReLU. It tries to solve the vanishing gradient problem. In normal ReLU our values become zero when input is neagtive this means that the gradient is zero which means that weights will never be updated so the neurons will not learn anything and we say that neurons are dead. That's why sometimes we want to use leak ReLU so whenever we notice that our weights won't update in our training then we try to use Leaky ReLU instead of ReLU

```python
import torch 
import torch.nn as nn
import torch.nn.functional as F
```

```python
x = torch.tensor([-1.0,1.0,2.0,3.0])
```

```python
#softmax
output = torch.softmax(x, dim=0)
print(output)

sm = nn.Softmax(dim=0)
output = sm(x)
print(output)
```

```python
#sigmoid 
output = torch.sigmoid(x)
print(output)

s = nn.Sigmoid()
output = s(x)
print(output)
```

```python
#tanh
output = torch.tanh(x)
print(output)


t = nn.Tanh()
output = t(x)
print(output)
```

```python
#relu
output = torch.relu(x)
print(output)

relu = nn.ReLU()
output = relu(x)
print(output)
```

```python
#leaky relu 

#sometimes some functions are not available in the torch API directly but 
#they are available in torch.nn.functional so we can import it as import torch.nn.functional as F

output = F.leaky_relu(x)
print(output)

lrelu = nn.LeakyReLU()
output = lrelu(x)
print(output)

```

```python
#option 1 (create nn modules)

class NeuralNet(nn.Module):
    def __init__(self, input_size,hidden_size):
        super(NeuralNet, self).__init__()
        self.linear1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(hidden_size,1)
        self.sigmoid = nn.Sigmoid()


    def forward(self, x):
        out = self.linear(x)
        out = self.relu(out)
        out = self.linear2(out)
        out = self.sigmoid(out)
        return(out)
```

```python
#option 2 (use activation functions directly in the forward pass)

class NeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(NeuralNet, self).__init__()
        self.linear1 = nn.Linear(input_size, hidden_size)
        self.linear2 = nn.Linear(hidden_size, 1)

    def forward(self, x):
        out = torch.relu(self.linear1(x))
        out = torch.sigmoid(self.linear2(out))
        return out

        
```
